```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/migracion.csv')
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Valores perdidos por columna:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
print("Valores perdidos por columna:")
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
# Detectar y filtrar outliers usando IQR para la columna 'Cantidad_Migrantes'

dfcopia = df.copy()
Q1 = dfcopia['Cantidad_Migrantes'].quantile(0.25)
Q3 = dfcopia['Cantidad_Migrantes'].quantile(0.75)
IQR = Q3 - Q1
lim_superior = Q3 + 1.5 * IQR
lim_inferior = Q1 - 1.5 * IQR

outliers = dfcopia[(dfcopia['Cantidad_Migrantes'] < lim_inferior) | (dfcopia['Cantidad_Migrantes'] > lim_superior)]

dfcopia_sin_outliers = dfcopia[(dfcopia['Cantidad_Migrantes'] > lim_inferior) & (dfcopia['Cantidad_Migrantes'] < lim_superior)]


print("Valores Q1 y Q3:")
print(Q1, Q3)

print("Limites superior e inferior:")
print(lim_superior, lim_inferior)

print("Valores atípicos (outliers) en 'Cantidad_Migrantes':")
print(outliers)
dfcopia_sin_outliers
```

```python
# Reemplazar valores en "Razon_Migracion"
mapeo_razon = {
    "Económica": "Trabajo",
    "Conflicto": "Guerra"
}
dfcopia_sin_outliers["Razon_Migracion"] = dfcopia_sin_outliers["Razon_Migracion"].replace(mapeo_razon)
dfcopia_sin_outliers
```

```python
# Media y mediana de cantidad de migrantes
media_migrantes = dfcopia_sin_outliers['Cantidad_Migrantes'].mean()
mediana_migrantes = dfcopia_sin_outliers['Cantidad_Migrantes'].median()
print(f"\nMedia de migrantes: {media_migrantes}")
print(f"Mediana de migrantes: {mediana_migrantes}")
```

```python
# PIB promedio de países de origen y destino
pib_promedio_origen = dfcopia_sin_outliers['PIB_Origen'].mean()
pib_promedio_destino = dfcopia_sin_outliers['PIB_Destino'].mean()
print(f"\nPIB promedio de origen: {pib_promedio_origen}")
print(f"PIB promedio de destino: {pib_promedio_destino}")
```

```python
# Contar cuántos movimientos por cada razón
conteo_razones = df['Razon_Migracion'].value_counts()
print("\nConteo de migraciones por razón:")
print(conteo_razones)
```

```python
# Agrupar por Razon_Migracion y calcular suma total de migrantes
agrupado_suma = df.groupby('Razon_Migracion')['Cantidad_Migrantes'].sum()
print("\nTotal de migrantes por razón:")
print(agrupado_suma)
```

```python
# Media del IDH de los países de origen por tipo de migración
media_idh_origen = df.groupby('Razon_Migracion')['IDH_Origen'].mean()
print("\nMedia del IDH de origen por razón de migración:")
print(media_idh_origen)
```

```python
# Ordenar DataFrame por cantidad de migrantes de mayor a menor
df_ordenado = df.sort_values(by='Cantidad_Migrantes', ascending=False)
print("\nDataFrame ordenado por cantidad de migrantes (descendente):")
print(df_ordenado)
```

```python
# Filtrar solo migraciones por guerra
df_guerra = df[df['Razon_Migracion'] == 'Guerra']
print("\nMigraciones por conflicto (guerra):")
print(df_guerra)
```

```python
# Crear nueva columna Diferencia_IDH
df['Diferencia_IDH'] = df['IDH_Destino'] - df['IDH_Origen']
print("\nDataFrame con la nueva columna 'Diferencia_IDH':")
print(df.head())
```

```python
# Guardar el DataFrame final en CSV sin índice
df.to_csv("Migracion_Limpio.csv", index=False)
dffinal = pd.read_csv('/content/Migracion_Limpio.csv')
dffinal.head()
```
